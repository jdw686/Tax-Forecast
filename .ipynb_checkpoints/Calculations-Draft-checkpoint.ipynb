```python
from forecasting import sarimax_forecast
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
import warnings
warnings.simplefilter('ignore')

revenues = pd.read_csv('Final Tables by Tax Source/final_data_revenue.csv')
revenues = revenues.iloc[:, 1:]
revenues['Category'].unique()
```

```python
lohnsteuer = revenues[revenues['Category'] == 'Lohnsteuer']
lohnsteuer = lohnsteuer.set_index('Date')
```

```python
lohnsteuer['Rolling Total'] = lohnsteuer['Monthly Total'].rolling(12).sum()
```

```python
dftest = adfuller(lohnsteuer['Rolling Total'].pct_change().dropna(), autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput.round(4))
```

```python

lohnsteuer['Rolling Change Diff'] = lohnsteuer['Rolling Total'].diff()
lohnsteuer['Rolling Change %'] = lohnsteuer['Rolling Total'].pct_change()
lohnsteuer_forecast_diff = sarimax_forecast(lohnsteuer['Rolling Change Diff'], grid_search_parameters = 2)
lohnsteuer_prediction_diff = lohnsteuer_forecast_diff.cumsum()+lohnsteuer['Rolling Total'].iloc[-1]
lohnsteuer_forecast_pct = sarimax_forecast(lohnsteuer['Rolling Change %'], grid_search_parameters = 2)
lohnsteuer_prediction_pct = (1+lohnsteuer_forecast_pct).cumprod()*lohnsteuer['Current Year Total'].iloc[-1]
```

```python
lohnsteuer_forecast_diff.cumsum()+lohnsteuer['Rolling Total'].iloc[-1]
```

```python
lohnsteuer['Rolling Change Diff'][-12:].sum()+lohnsteuer['Rolling Total'].iloc[-13]
```

```python
(1+lohnsteuer_forecast_pct).cumprod()*lohnsteuer['Current Year Total'].iloc[-1]
```

```python
revenues[revenues['Month'] == 12].groupby(['Year', 'Category'])[['Current Year Total']].sum().unstack()
```

# Solidaritätszuschlag (Solidarity Surcharge)

```python
solidarity = revenues[revenues['Category'] == 'Solidaritätszuschlag']
solidarity = solidarity.set_index('Date')
solidarity['Rolling Total'] = solidarity['Monthly Total'].rolling(12).sum()
```

```python
dftest = adfuller(solidarity['Rolling Total'].diff().dropna(), autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput.round(4))
```

```python
solidarity['Rolling Change Diff'] = solidarity['Rolling Total'].diff()
solidarity['Rolling Change %'] = solidarity['Rolling Total'].pct_change()
months_diff = solidarity.groupby('Month')['Rolling Change Diff'].mean()
months_list_diff = list(months_diff.values)*11
solidarity['Average Rolling Change Diff'] = solidarity['Rolling Change Diff']/months_list_diff
months_pct = solidarity.groupby('Month')['Rolling Change %'].mean()
months_list_pct = list(months_pct.values)*11
solidarity['Average Rolling Change %'] = solidarity['Rolling Change %']/months_list_pct
```

```python
solidarity_forecast_diff = sarimax_forecast(solidarity['Average Rolling Change Diff'], grid_search_parameters = 2)
solidarity_forecast_pct = sarimax_forecast(solidarity['Average Rolling Change %'], grid_search_parameters = 2)
```

```python
solidarity_prediction_diff = months_list_diff[:12]*solidarity_forecast_diff.cumsum()+solidarity['Rolling Total'].iloc[-1]
solidarity_prediction_pct = (1+months_list_pct[:12]*solidarity_forecast_pct).cumprod()*solidarity['Current Year Total'].iloc[-1]
```

```python
pd.DataFrame({'Soli Rolling Total Diff': solidarity_prediction_diff, 
              'Soli Rolling Total %': solidarity_prediction_pct})
```

```python
df = revenues[revenues['Category'] == 'Lohnsteuer']
df = df.set_index('Date')
df['Rolling Total'] = df['Monthly Total'].rolling(12).sum()

######


def forecast(df, steps = 12):
    test1 = pd.Series(adfuller(df['Rolling Total'].dropna(), autolag='AIC')[0:4], 
                         index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    test2 = pd.Series(adfuller(df['Rolling Total'].pct_change().dropna(), autolag='AIC')[0:4], 
                         index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    test3 = pd.Series(adfuller(df['Rolling Total'].diff().dropna(), autolag='AIC')[0:4], 
                         index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])

    df_forecast2 = sarimax_forecast(df['Rolling Total'].diff(), periods = steps)
    forecast2_series = df_forecast2.cumsum()+df['Rolling Total'].iloc[-1] if test2[1] < 0.05 else pd.Series(np.nan)

    df_forecast3 = sarimax_forecast(df['Rolling Total'].pct_change(), periods = steps)
    forecast3_series = (1+df_forecast3).cumprod()*df['Current Year Total'].iloc[-1] if test3[1] < 0.05 else pd.Series(np.nan)

    forecast_df = pd.concat([pd.Series(df['Category'][-1], index = forecast3_series.index), forecast2_series, forecast3_series], axis =1)
    forecast_df.columns = ['Category', 'Difference Method', 'Percentage Method']
    
    return forecast_df
```
