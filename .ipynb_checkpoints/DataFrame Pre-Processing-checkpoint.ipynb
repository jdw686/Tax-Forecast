```python
tax_source = ['Lohnsteuer', 
              'Veranlagte Einkommensteuer', 
              'nicht veranlagte Steuer vom Ertrag', 
              'Abgeltungsteuer auf Zins- und Veräußerungserträge',
              'Körperschaftsteuer',
              'Steuern vom Umsatz',
              'Gewerbesteuerumlage',
              'Energiesteuer',
              'Tabaksteuer',
              'Solidaritätszuschlag',
              'Versicherungsteuer',
              'Stromsteuer',
              'Kraftfahrzeugsteuer',
              'Kernbrennstoffsteuer',
              'Kaffeesteuer',
              'Alkoholsteuer',
              'Luftverkehrsteuer',
              #different loop needed to create negative numbers
              'Ergänzungszuweisungen an Länder',
              'BNE-Eigenmittel der EU',
              'Mehrwertsteuer-Eigenmittel der EU',
              'Zuweisungen an Länder für ÖPNV',
              'Zuweisung an die Länder für Kfz-Steuer und Lkw-Maut',
              'Sonstige Einnahmen']

deductions = ['Ergänzungszuweisungen an Länder',
              'BNE-Eigenmittel der EU',
              'Mehrwertsteuer-Eigenmittel der EU',
              'Zuweisungen an Länder für ÖPNV',
              'Zuweisung an die Länder für Kfz-Steuer und Lkw-Maut',
              'Sonstige Einnahmen']

#recreates the 
for i in deductions:
    df = categories[categories['Category'] == f'{i}'].reset_index().drop('index', axis = 1)
    df['Date'] = pd.to_datetime(df['Date'])
    df_2008 = df[df['Year'] == 2009]
    df_2008['Year'] = 2008
    df_2008['Current Year Total'] = df_2008['Prior Year Total']
    df_2008['Prior Year Total'] = 0
    df_2008['Date'] = df_2008['Year'].astype(str).str.cat(df_2008['Month'].astype(str), sep = '-')
    df_2008['Date'] = pd.to_datetime(df_2008['Date'])
    df = pd.concat([df_2008, df])
    df['Prior Year Total'] = df['Prior Year Total'].where(
        df['Prior Year Total'] <= 0, other = df['Prior Year Total']*-1)
    df['Current Year Total'] = df['Current Year Total'].where(
        df['Current Year Total'] <= 0, other = df['Current Year Total']*-1)
    df['Monthly Total'] = df['Current Year Total'].where(
        df['Month'] == 1).fillna(df['Current Year Total'] - df['Current Year Total'].shift())
    df.to_csv(f'{i}.csv')
# df
```
