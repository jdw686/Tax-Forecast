```python
import numpy as np
import re

def remove_x_space(s):
    try:
        new = re.sub('([\xa0])', '0',s)
    except TypeError:
        new = np.nan
    except ValueError:
        new = np.nan
    return new

def replace_pd(s):
    try:
        new = re.sub('([\.])', '',s)
    except TypeError:
        new = np.nan
    except ValueError:
        new = np.nan
    return new

def remove_white_space(s):
    try:
        new = re.sub('([\s])', '',s)
    except TypeError:
        new = np.nan
    except ValueError:
        new = np.nan
    return new

def table_converter_new(url):
    test = html_tables(url)
    table = test.read()
    df = pd.DataFrame(table[-1][[0, 6]])
    df[0] = df[0].apply(remove_n_space)
    df[6] = df[6].apply(remove_x_space).apply(replace_pd).apply(remove_white_space)
    df[2] = int(input('Insert Year Here: '))
    df[3] = int(input('Insert Month Here: '))
    df = df.dropna()
#     df.to_csv(f'{year}-{month}.csv')
    return df

def table_converter_all(url):
    test = html_tables(url)
    table = test.read()
    df = pd.DataFrame(table[-1])
    df[0] = df[0].apply(remove_n_space)
    df[1] = df[].apply(remove_x_space).apply(replace_pd).apply(remove_white_space)
    df[5] = df[5].apply(remove_x_space).apply(replace_pd).apply(remove_white_space)
    df['Year'] = int(input('Insert Year Here: '))
    df['Month'] = int(input('Insert Month Here: '))
    df = df.dropna()
#     df.to_csv(f'{year}-{month}.csv')
    return df
```
